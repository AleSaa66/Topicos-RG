### Semana 5: Geodésicas tipo luz - desvío de luz

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('bmh')
```

La solución perturbativa a primer orden de la EDO
$$
u'' + u = 3mu^2,
$$
para $u := 1/r$, es
$$
u(\varphi) = \frac{1}{D} \left[ \sin(\varphi - \varphi_0) + \frac{m}{D} \left[ 1 + \cos^2(\varphi-\varphi_0)\right] \right] + \mathcal{O}(\epsilon^2).
$$

```python
#Cálculo del parametro m para el Sol:
G = 6.67430e-11         #[m^3 kg^-1 s^-2]
M = 1.988e30            #[kg]
c = 3e8                 #[m s^-1]

m = G*M/c**2 
print("El parametro m =",m)
```

```python
#Valores de epsilon y cálculo de la distancia D:
ep = np.array([0.0001,0.0003,0.0005,0.0007])
D = 3*m/ep

print("epsilon =",ep,"y D =",D)
print("El primer valor de D es D_0=",D[0])
#Nota: El parametro m se mantiene fijo, pero se varía epsilon, lo cual se traduce a una variación de la distancia del Sol y del rayo de luz.
```

```python
#Partición de la variable angular phi:
k = 0.5
phi = np.linspace(0,2*k*np.pi,500)

#Cálculo de la función u adimensionalizada con respecto al valor de D[0]:
u_list = []

for i in range(len(ep)):
    u_0 = (np.sin(phi) + (ep[i]/3)*(1 + np.cos(phi)**2))*(D[0]/D[i])
    u_list.append(u_0)

u = np.array(u_list)
```

```python
#Cálculo de la coordenada radial:
r = 1/u

#Variable phi bidimensional con una copia de phi en cada fila según len(r):
phi_2d = np.multiply.outer(np.ones(len(r)),phi)

#Polares a cartesianas:
x = r*np.cos(phi_2d)
y = r*np.sin(phi_2d)
```

```python
#Gráfica del desvío de la luz para diferentes valores de epsilon:
colors = ['red','blue','green','orange']
for i in range(len(r)):
    plt.plot(x[i],y[i], color = colors[i], linewidth = 1.3, label = "$\\epsilon =$"+str(ep[i]))

#Gráfica de la luz caso newtoniano:
for j in range(len(r)):
    plt.hlines(D[j]/D[0],xmin = -100,xmax = 100,color = 'gray',linestyle = '--',linewidth = 1.3)

#Sol:
plt.scatter(0,0,s = 3000.0)

plt.grid(True)
plt.xlim(-100,100)
plt.ylim(0.0,1.05)

plt.xlabel('$x/D_0$',fontsize = 15)
plt.ylabel('$y/D_0$',fontsize = 13)
plt.legend(loc = 'best')
```
